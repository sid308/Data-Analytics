```python
import pandas as pd
import matplotlib.pyplot as plt

days = ["day1", 'day2', 'day3', "day4", "day5"]
value = [300,420,250,230,400]
y1 = [500,450,300,250,320]
plt.plot(days,value,marker = "o",ls="-",color="red",label='week1',alpha=0.3)
plt.plot(days,y1,marker = "o",ls="--",color="green",label='week2')
plt.legend()
plt.show()
```
