```python
import matplotlib.pyplot as plt
import pandas as pd

rating = [98,67,88,95, 88]
movie = ["Part1", "Part2", "Part3", "Part4", "Part5"]
colors = ["red",'green','yellow','blue','purple']

plt.bar(movie,rating,color=colors,edgecolor='black')
plt.xlabel("Movie Part",fontsize=17)
plt.ylabel("Rating",fontsize=17)
plt.title("Harry Puttar",fontsize=20)
plt.show()
```

```python
data = pd.read_excel("employees.xlsx")
df = pd.DataFrame(data)
print(df.head(10))
grouped_by = df.groupby(['Department'])["Annual Salary"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.show()
```
