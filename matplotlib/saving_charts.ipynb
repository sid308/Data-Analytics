```python
import matplotlib.pyplot as plt

brands = ["OnePlus","Apple","Samsung","Nokia","Redmi"]
popularity = [22,35,30,12,10]
color = ["red","purple","blue","magenta","orange"]
ex = [0,0,0,0,0.1]

plt.pie(popularity,labels=brands,colors=color,explode=ex,shadow=True,autopct="%.2f",startangle=90)
plt.savefig("pie.png",facecolor="pink",pad_inches=0.9,bbox_inches="tight")
plt.show()
```
