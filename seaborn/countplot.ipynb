```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
colors = ['blue','green','orange','red']
print(data.head(1))

sns.countplot(data=data,x='day',hue='day')
plt.show()
```
