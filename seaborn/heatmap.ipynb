```python
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
gp = data.groupby('day',observed=True).agg({'tip':'mean'})
print(gp)
sns.heatmap(gp)
```

```python
data = pd.read_excel("employees.xlsx")
gp = data.groupby("Job Title").agg({'Annual Salary':'mean'})
sns.heatmap(gp)
```
